```kotlin
%use lets-plot

//Historgam chart

import kotlin.math.abs
import java.util.Random

val rand = Random()

val data = mapOf(
"x" to List(500) {rand.nextGaussian()} + List(500) {rand.nextGaussian()},
"type" to List(500) {"A"} + List(500) {"B"}
)

ggplot(data) + geomHistogram(alpha = 0.6 , size = 0.0) {x = "x"; fill = "type"}
```
